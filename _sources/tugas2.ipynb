```python
import numpy as np
import numpy as sy
```

```python
A = np.array([[1,3,9],[3,-2,6],[9,4,1]])
B = np.array([[1,7,5],[2,4,5],[2,4,1]])
C = np.array([[3,2],[-1,2],[3,5]])
D = np.array([[3,4,7],[2,7,3],[4,1,3]])
E = A+B
F = 2*A
G = A@C
H = B@C
I = A@H
J = A@B
K = J@C

print(A)
print('\n')
print(B)
print('\n')
print(C)
print('\n')
print(D)
print('\n')
print(E)
print('\n')
print(F)
print('\n')
print(G)
print('\n')
print(H)
print('\n')
print(I)
print('\n')
print(J)
print('\n')
print(K)
print('\n')
```
